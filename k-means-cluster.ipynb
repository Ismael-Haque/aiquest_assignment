```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mall_customers = pd.read_csv("mall customers.csv")
mall_customers.head()
```

```python
mall_customers = mall_customers.drop("CustomerID", axis=1)
mall_customers.head()
```

```python
mall_customers = mall_customers.rename(columns = {'Gender': 'gender', 'Age':'age', 'Annual Income (k$)':'income', 'Spending Score (1-100)':'score'})
mall_customers.head()
```

```python
mall_customers.describe()
```

```python
plt.scatter(mall_customers.age, mall_customers.income)
```

```python
plt.scatter(mall_customers.score, mall_customers.income, color="red")
plt.xlabel("Score in (1 to 100)")
plt.ylabel("Income")
plt.title("Age and Score plot")
```

```python
sns.pairplot(mall_customers[['age', 'income', 'score']])
```

```python
plt.scatter(mall_customers.score, mall_customers.income, color="black")
plt.xlabel('Score in (1 to 100)')
plt.ylabel("Income")
plt.title('Age and Score plot')
```
